Titanic Analytics using SQL

```python
import pandas as pd
import pandasql as ps
```

```python
titanic_df = pd.read_csv('/Users/isingh/Desktop/titanic.csv')
```

```python
titanic_df.head(5)
```

SELECTING ALL ROWS ON DATASET

```python
ps.sqldf("""SELECT 
    *
    FROM titanic_df
""")
```

SELECTING ONLY SPECIFIC COLUMNS

```python
ps.sqldf("""
SELECT df.Name, df.Sex, df.Age FROM titanic_df AS df LIMIT 10 """)
```

RENAMING COLUMNS

```python
ps.sqldf("""
SELECT df.Name AS name_of_passenger,df.Sex AS sex_of_passenger,df.Age AS age_of_passenger FROM titanic_df AS df LIMIT 10""")
```

CONDITIONALS --> Where age is 35 and sex is female

```python
ps.sqldf("""
SELECT df.* FROM titanic_df AS df
WHERE (df.Age=35 AND df.Sex="female")""")
```

MISSING VALUES SEARCH

```python
ps.sqldf("""
SELECT df.* FROM titanic_df AS df
WHERE df.age IS NULL""")
```

FILTERING WITH NAMES IN BETWEEN

```python
ps.sqldf("""
SELECT df.* FROM titanic_df AS df
WHERE df.name LIKE "%Mrs.%"
LIMIT 15
""")
```

CREATING CATEGORIES

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Sex,
    df.Age,
    CASE WHEN df.Age IS NULL THEN 'Missing'
        WHEN df.Age <18 THEN '1-17'
        WHEN df.Age <60 THEN '18-60'
    ELSE '3. 60+' END AS age_bucket
FROM titanic_df AS df
LIMIT 50   
    
""")
```

AGGREGATION

```python
ps.sqldf("""
SELECT 
df.sex,
AVG(df.Age) AS avg_age,
MAX(df.AGE) AS max_age,
COUNT(df.PassengerId) AS count_passengers,
COUNT(DISTINCT df.Ticket) AS count_tickets
FROM titanic_df AS df
GROUP BY df.sex
""")
```
